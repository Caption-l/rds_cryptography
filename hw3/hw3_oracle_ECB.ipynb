```python
from Crypto.Cipher import AES
from binascii import hexlify
import requests, json, string


def encrypt(pt):
    """Отримуємо шифртекст (шифрування) для відкритого тексту"""
    hex_pt = hexlify(pt.encode()).decode()
    url = f"http://aes.cryptohack.org/ecb_oracle/encrypt/{hex_pt}"
    r = requests.get(url)
    return json.loads(r.text)["ciphertext"]

def find_flag():
    known_flag = '' #Знайдена частина FLAG
    total = 31 #Для аналізу використовуємо 2 блоки по 16 байт (але порівняння буде лише по блоку 2)
    alphabet = string.printable #Підбір літер - по друкованих символах

    while True:
        prefix = '-' * (total-len(known_flag)) #еталонний відкритий текст
        expected = encrypt(prefix)             #еталонний зашифрований текст

        # Тепер у циклі перебираємо друковані символи з метою перевірки
        # на співпадіння з еталонним шифротекстом
        for char in alphabet:
            res = encrypt(prefix + known_flag + char) #рядок для порівняння

            if res[32:64] == expected[32:64]: #порівняння йде по блоку 2
                known_flag += char            #додаємо знайдений символ
                print(known_flag)
                break

        #оскільки знаємо наперед формат FLAG, то критерієм зупинки буде символ '}'
        if known_flag.endswith('}'): break

    print(known_flag)

find_flag()
```
