```python
import requests
import json
from binascii import hexlify
```

```python
def get_cookie():
  #Obtain cookie (iv+dc)'
  url = "http://aes.cryptohack.org/flipping_cookie/get_cookie/"
  r = requests.get(url)
  ct = (json.loads(r.text))['cookie']
  return ct

def check_admin(cookie, iv):
  #Check if admin. If admin, return FLAG otherwise return error
  url = "http://aes.cryptohack.org/flipping_cookie/check_admin/" + cookie + "/" + iv
  r = requests.get(url)
  ct = (json.loads(r.text))
  return ct

#Get any cookie
cookie0 = get_cookie()
print ('cookie_false:', cookie0)

#Split cookie into iv0 and dc0
iv0 = cookie0[:32]
dc0 = cookie0[32:]

print('iv0:', iv0)
print('dc0:', dc0)

#Get hex("admin=True;00000")
p1_text = "admin=True;00000"
p1_hex = hexlify(p1_text.encode("utf-8")).decode()

#Get hex("admin=False;00000")
p0_text = "admin=False;0000"
p0_hex = hexlify(p0_text.encode("utf-8")).decode()

print('p1_hex:', p1_hex)
print('p0_hex:', p0_hex)

#Get IV'
iv1 = hex(int(iv0, 16) ^ int(p1_hex, 16) ^ int(p0_hex, 16))[2:]
print('iv1', iv1)

#Check if admin and receive FLAG
print(check_admin(dc0, iv1))
```
