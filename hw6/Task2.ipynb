```python
from cryptography.hazmat.primitives.asymmetric import padding, rsa
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives import serialization

path = "/content/drive/MyDrive/robot_dreams_cryptography/HW6/"
```

```python

# Завантаження відкритого ключа
with open(path + "task_pub.pem", "rb") as key_file:
    public_key = serialization.load_pem_public_key(key_file.read())

# Завантаження повідомлення
with open(path + "task2_plain_message.txt", "r", encoding="utf-8") as msg_file:
    message = msg_file.read().encode("utf-8")

# Зашифрування повідомлення за допомогою RSA-OAEP
ciphertext = public_key.encrypt(
    message,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)

# Збереження зашифрованого повідомлення у форматі hex
with open(path + "task-2-message.txt", "w") as enc_file:
    enc_file.write(ciphertext.hex())

print("Повідомлення успішно зашифровано та збережено у task-2-message.txt")
```
