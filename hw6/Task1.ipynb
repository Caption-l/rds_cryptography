```python
from cryptography.hazmat.primitives.asymmetric import padding, rsa
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives import serialization

path = "/content/drive/MyDrive/robot_dreams_cryptography/HW6/"
```

```python
# Завантаження відкритого ключа
with open(path + "task_pub.pem", "rb") as key_file:
    public_key = serialization.load_pem_public_key(key_file.read())

# Завантаження повідомлення
with open(path + "task_message.txt", "r") as msg_file:
    message_hex = msg_file.read()

message = bytes.fromhex(message_hex)              #конвертація рядку із hex-формату у байти
print(f"Повідомлення: {message.decode('utf-8')}") #вивід змісту рядка на екран

# Завантаження підпису
with open(path + "task_signature.txt", "r") as sign_file:
    signature = bytes.fromhex(sign_file.read())

# Перевірка підпису
try:
    public_key.verify(
        signature,
        message,
        padding.PSS(
            mgf=padding.MGF1(hashes.SHA256()),
            salt_length=padding.PSS.MAX_LENGTH
        ),
        hashes.SHA256()
    )
    print("Підпис валідний")
except Exception as e:
    print("Підпис недійсний")
```
