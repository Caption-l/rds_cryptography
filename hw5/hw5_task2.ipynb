```python
from Crypto.Hash import HMAC, SHA256
from Crypto.Protocol.KDF import HKDF
from Crypto.Random import get_random_bytes

# Вхідні дані
master_key = bytes.fromhex("63e353ae93ecbfe00271de53b6f02a46")  # 128-бітний ключ
iv = bytes.fromhex("75b777fc8f70045c6006b39da1b3d622")
ciphertext = bytes.fromhex("76c3ada7f1f7563ff30d7290e58fb4476eb12997d02a6488201c075da52ff3890260e2c89f631e7f919af96e4e47980a")

# Генеруємо ключ для автентифікації (для MAC) (32 байти для HMAC-SHA256)
salt = get_random_bytes(32) # створюємо сіль як випадкове 32-байтне значення; сіль необхідно зберегти
key_mac = HKDF(master_key, 32, salt=salt, hashmod=SHA256, num_keys=1) #формуємо 32-байтний ключ автентифікації

# Формуємо MAC (використовуючи алгоритм HMAC-SHA256)
hmac = HMAC.new(key_mac, digestmod=SHA256)
hmac.update(iv)
hmac.update(ciphertext)
mac_tag = hmac.digest()

# Виводимо MAC у HEX
print("MAC:", mac_tag.hex())
```
