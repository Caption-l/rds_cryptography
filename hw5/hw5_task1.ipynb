```python
from argon2 import PasswordHasher, Type

ph = PasswordHasher(
    time_cost=2,          # кількість ітерацій (trade-off між продуктивністю і безпекою)
    memory_cost=65536,    # пам'ять у КБ (64 MB)
    parallelism=4,        # паралельність (припускаємо, що на сервері є мінімум 4 ядерні процесори)
    hash_len=32,          # довжина отриманого хешу
    salt_len=16,          # довжина солі
    type=Type.ID          # визначає Argon2id, рекомендовану версію з найвищою стійкістю серед Argon2
)
```
